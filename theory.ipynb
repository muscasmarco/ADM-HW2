We also wrote a script to test the behaviour of the function and compare it to the expected graph of a function we found to be the result.

```python
import time
import matplotlib.pyplot as plt
import numpy as np
```


Here are the function definitions according to what was posted on Github.

```python
def splitSwap(a, i, n):

    if n <= 1:
        return 
    n_half = int(n/2)
    splitSwap(a, i, n_half)
    splitSwap(a, i+n_half, n_half)
    
    swapList(a, i, n)

def swapList(a, l, n):

    n_half = int(n/2)
    for i in range(0, n_half):
        tmp = a[l + i]
        a[l + i] = a[l + n_half + i]
        a[l + n_half + i] = tmp    
```

Here is just a small test for array sizes expressible as 2^k. 
Non ideal case can be tested changing the value of n inside the for loop.

```python
array_sizes = []
times = []
x_squared = []
log = []

for i in range(1,22):
    n = 2 ** i
    array = list(range(1, n+1))

    start_time = time.time()
    splitSwap(array, 0, n)
    end_time = time.time() - start_time

    array_sizes.append(str(n))
    times.append(end_time)
    x_squared.append(n ** 2)
    log.append(n * np.log2(n))
```

We will now compare the different graphs obtained from the data at the previous step.

Here is the execution time according to the input size.

```python
''' Execution time '''
plt.figure(figsize=(10, 4))
plt.plot(times, color='orange')  
plt.xlabel('n')
plt.ylabel('execution time (s)')
plt.xticks(np.arange(0, len(array_sizes)), array_sizes, rotation='vertical')
plt.show()
```

Here the Y axis is not expressed in a particular unit. It is just for comparison: when we get an input 
    size of n, whats the plot of the function n log2(n)?

```python
''' f(x) = n log2(n) '''
plt.figure(figsize=(10, 5))
plt.plot(log, color='blue')  
plt.xlabel('n')
plt.ylabel('f(n) = n * log(n) base 2')
plt.xticks(np.arange(0, len(array_sizes)), array_sizes, rotation='vertical')
plt.show()
```

The function looks exponential, let's compare it for f(n) = n^2 

```python
plt.figure(figsize=(10, 5))
plt.plot(x_squared, color='blue')  
plt.xlabel('n')
plt.ylabel('f(n) = x^2')
plt.xticks(np.arange(0, len(array_sizes)), array_sizes, rotation='vertical')
plt.show()  
```

The function looks similar, but the slope change is greater. Also the f(n) of x^2 is greater by several orders of magnitude.
